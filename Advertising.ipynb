# Learning Algorithm

```python
from sklearn.linear_model import LinearRegression
new=['TV','Radio','Newspaper']

X=df[new]
Y=df['Sales']
```

```python
model=LinearRegression()
model=model.fit(X,Y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
Y_pred=model.predict(X)
Y_pred
```

```python
# score
model.score(X,Y)
```

```python
# give the input and predict the sales
new_x=[[600,300,60]]
model.predict(new_x)
```

```python
# predict the score
new=[[230.1,37.8,69.2]]
model.predict(new)
```
